# Testing on 200 images

At this step we will define two functions- choose_folder() and image_path() which heps us by, taking two 
random images from the dataset and testing it.

```python
def choose_folder():
    folder_label=["00001","00002","00003","00004","00005","00006","00007",
                   "00008","00009","00010","00011","00012","00013","00014",
                    "00015","00016","00017","00018","00019","00020","00021",
                     "00022","00023","00024","00025","00026","00027","00028",
                     "00029","00030","00031","00032","00033","00034","00035",
                      "00036","00037","00038","00039","00040","00041","00042"]
    
    folder=random.choice(folder_label)
    return folder


def image_path():
    folder=choose_folder()
    print("****************************************************************************")
    print("the True Label is",int(folder))
    path = r"/home/pulkit/Downloads/GTSRB/Training/"
    path=path+folder+"/"
    #print("the path is ",path)
    my_image = random.choice([
    x for x in os.listdir(path)
    if os.path.isfile(os.path.join(path, x))
    ])
    p=os.path.join(path, my_image)
    return p,folder
```

At this step, we will define the fuctions which will be important to process the image.

```python
def grayscale(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    return img

def equalize(img):
    img = cv2.equalizeHist(img)
    return img

def preprocessing(img):
    img = grayscale(img)
    img = equalize(img)
  #normalize the images, i.e. convert the pixel values to fit btwn 0 and 1
    img = img/255
    return img
```

At this step, we will define a fuction random_fuzzing_model_3, which will load the image, model, manipulate 
random 280 pixels and check the accuracy of the model on the manipulated images.

```python
def random_fuzzing_model_3():
    # loading the model
    model = tf.keras.models.load_model('/home/pulkit/Desktop/trained/noLoss_model.h5')
    
    y_true=[]
    y_pred=[]
    
    for i in range(0,200,1):
        li_x=[]
        li_y=[]
        for i in range(0,70,1):  
            
            li_x.append(random.randint(0,40))
            li_y.append(random.randint(0,40))
           
        # loading the image
        path,label=image_path()
        y_true.append(int(label))
        img = cv2.imread(path,1)
        
        # manipulating random pixels (70*4=280 pixels manipulated in 70 iterations on image)
        for i in range(0,70,1):
            
            img[li_x[i]:li_x[i]+4,li_y[i]:li_y[i]+1] = (255,255,255)
        
        # preprocessing the image so that it can be given an input to the model
        img= tf.cast(img, tf.float32)/255
        img = tf.image.resize(img, (32, 32))
        img = tf.expand_dims(img , 0)
        
        # predicting the labels
        pred=int(model.predict_classes(img))
        y_pred.append(pred)
        
        print("The Predicted label is "+ str(model.predict_classes(img)))
       
    # getting the accuracy score
    
    print("****************************************************************************")
    correct_percent=accuracy_score(y_true, y_pred)
    print(correct_percent)
```

```python
random_fuzzing_model_3()
```

Conclusion-: Applying Random Fuzzing decreases the prediction accuracy of the model from 93% to 44%. 
In order to further decresase the the prediction accuracy, we can apply the Random Fuzzing by manipulating
more pixels. Further applying Random Fuzzing on the image will make the quality of the image so bad, that 
even the human eye will not be able to recognize it.
